```python
from sklearn.datasets import make_regression
import numpy as np
import matplotlib.pyplot as plt
```

```python
X, y = make_regression(n_samples=4, n_features=1, n_informative=1, n_targets=1, noise=80,random_state=13)
plt.scatter(X, y)
```

```python
X
```

```python
y
```

```python
from sklearn.linear_model import LinearRegression

reg = LinearRegression()
reg.fit(X, y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
plt.scatter(X, y)
plt.plot(X, reg.predict(X), c='Red')
```

```python
# lets apply gradient descent slop is constant and its 78.35
# lets start with a b value b = 0 (intercept)

y_pred = ((78.35 * X) + 0).reshape(4)
```

```python
y_pred
```

```python
plt.scatter(X, y)
plt.plot(X, reg.predict(X), c='Red', label='OLS')
plt.plot(X, y_pred, c='green', label='b = 0')
plt.legend()
```

```python
b = 0
m = 78.35

loss_slope = -2 * np.sum(y - m * X.ravel() - b)
loss_slope
```

```python
lr = 0.1

step_size = lr * loss_slope
step_size
```

```python
b = b - step_size
```

```python
y_pred_new = m * X + b
y_pred_new
```

```python
m = 78.35
lr = 0.1
epoch=10

def gradient(x, y):
    b=100
    for i in range(epoch):
        loss_slope = -2 * np.sum(y - m * x.ravel() - b)
        b = b - (lr * loss_slope)
        
        y_pred = m * x + b
        plt.plot(x, y_pred)
    return b

b_final = gradient(X, y)
b_final
```
